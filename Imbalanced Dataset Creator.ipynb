# Getting Started

```python
import shutil
import os
import random
```

```python
src = r"Dataset Folder\Original\Concrete Crack Images for Classification"
parent_dst = r"Dataset Folder\Actual"

dst = os.path.join(parent_dst, os.path.basename(src))
shutil.copytree(src, dst, dirs_exist_ok=True)
```

```python
random.seed(2003)
```

```python
cracked_folder = r"Dataset Folder\Actual\Concrete Crack Images for Classification\Positive"
unseen_folder = r"Dataset Folder\Actual\Concrete Crack Images for Classification\Unseen"

os.makedirs(unseen_folder, exist_ok=True)

gitkeep_src = os.path.join(cracked_folder, ".gitkeep")
gitkeep_dst = os.path.join(unseen_folder, ".gitkeep")

if os.path.isfile(gitkeep_src) and not os.path.exists(gitkeep_dst):
    shutil.copy2(gitkeep_src, gitkeep_dst)
```

```python
delete_percentage = 0.8
dry_run = False

cracked_files: list[str] = []
for file in os.listdir(path=cracked_folder):
	if file == ".gitkeep":
		print(".gitkeep ignored")
		continue

	image_path = os.path.join(cracked_folder, file)

	if os.path.isfile(image_path):
		cracked_files.append(file)
```

```python
total = len(cracked_files)
delete_count = int(total * delete_percentage)

to_delete = random.sample(cracked_files, delete_count)

if not dry_run:
	for file in to_delete:
		src_path = os.path.join(cracked_folder, file)
		dst_path = os.path.join(unseen_folder, file)
		shutil.move(src_path, dst_path)
	print(f"{delete_count} files moved to {unseen_folder}")
else:
	print(f"Dry run, no files moved. {delete_count} files would be moved to {unseen_folder}")
```
