```python
import os
import random
```

```python
random.seed(2003)
```

```python
cracked_folder = "./Images/Positive"

delete_percentage = 0.8
dry_run = False

cracked_files: list[str] = []
for file in os.listdir(path=cracked_folder):

	image_path = os.path.join(cracked_folder, file)

	if os.path.isfile(image_path):
		cracked_files.append(file)
```

```python
total = len(cracked_files)
delete_count = int(total * delete_percentage)

to_delete = random.sample(cracked_files, delete_count)

if not dry_run:
	for file in to_delete:
		image_path = os.path.join(cracked_folder, file)
		os.remove(image_path)
	print(f"{delete_count} files deleted")
else:
	print("Dry run, no files deleted")
```
